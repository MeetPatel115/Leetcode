```python
def calculate(s):
    s = s.replace(" ", "")
    
    def apply_operator(op, left, right):
        if op == '*':
            return str(int(left) * int(right))
        elif op == '/':
            return str(int(left) / int(right))
        elif op == '-':
            return str(int(left) - int(right))
        elif op == '+':
            return str(int(left) + int(right))

    operators = ['*', '/', '+', '-']
    for op in operators:
        while op in s:
            for j in range(1, len(s) - 1):
                if s[j] == op:
                    left = s[j - 1]
                    right = s[j + 1]
                    result = apply_operator(op, left, right)
                    s = s[:j - 1] + result + s[j + 2:]
                    break

    return s

# Example usage
result = calculate('3+2*2')
print(result)  # Output should be '6'
```

```python
class Solution:
    def calculate(self, s: str):
        if not s:
            return 0
        
        stack, curr_num, operator = [], 0, "+"
        all_operators = {"+", "-", "*", "/"}
        nums = set(str(x) for x in range(10))
        
        for indx in range(len(s)):
            char = s[indx]
            
            if char in nums:
                curr_num = curr_num * 10 + int(char)
            
            if char in all_operators or indx == len(s) - 1:
                if operator == "+":
                    stack.append(curr_num)
                    
                elif operator == "-":
                    stack.append(-curr_num)
                    
                elif operator == "*":
                    stack[-1] *= curr_num
                
                elif operator == "/":
                    stack[-1] = int(stack[-1] / curr_num)
                    
                curr_num = 0
                operator = char
            
        return sum(stack)
    
```
